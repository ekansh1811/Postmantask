```python
from sklearn.datasets import make_circles, make_classification, make_moons
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score, accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import cross_validate
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
gs = GaussianNB()
```

```python
x_train.info(), x_test.info()
```

```python
gsinit = time.time()
for i in range(79):
    gs.partial_fit(x_train[i*10000:10000*(i+1)],y_train[i*10000:10000*(i+1)],classes = [0,1])
gst = time.time()-gsinit
```

```python
gs.score(x_test,y_test),gst
```

```python
kn = KNeighborsClassifier()
```

```python
kninit = time.time()
kn.fit(x_train,y_train)
knt = time.time()-kninit
```

```python
kn.score(x_test,y_test),knt
```

```python
nn = MLPClassifier(random_state=1,
                  solver = 'sgd',
                  batch_size = 400,
                  learning_rate_init = 0.005,
                  momentum = 0.5)
```

```python
nn.fit(x_train,y_train)
```

```python
nn.score(x_test,y_test)
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
dt.score(x_test,y_test)
```

```python
scoring = ['accuracy', 'precision_macro', 'recall_macro' , 'f1_weighted', 'roc_auc']
scores = cross_validate(gs, x_test, y_test, scoring=scoring, cv=20)
gs_fit_time = scores['fit_time'].mean()
gs_score_time = scores['score_time'].mean()
gs_accuracy = scores['test_accuracy'].mean()
gs_precision = scores['test_precision_macro'].mean()
gs_recall = scores['test_recall_macro'].mean()
gs_f1 = scores['test_f1_weighted'].mean()
gs_roc = scores['test_roc_auc'].mean()
scores
```

```python
scoring = ['accuracy', 'precision_macro', 'recall_macro' , 'f1_weighted', 'roc_auc']
scores = cross_validate(kn, x_test, y_test, scoring=scoring, cv=20)
kn_fit_time = scores['fit_time'].mean()
kn_score_time = scores['score_time'].mean()
kn_accuracy = scores['test_accuracy'].mean()
kn_precision = scores['test_precision_macro'].mean()
kn_recall = scores['test_recall_macro'].mean()
kn_f1 = scores['test_f1_weighted'].mean()
kn_roc = scores['test_roc_auc'].mean()
scores
```

```python
scoring = ['accuracy', 'precision_macro', 'recall_macro' , 'f1_weighted', 'roc_auc']
scores = cross_validate(nn, x_test, y_test, scoring=scoring, cv=20)
nn_fit_time = scores['fit_time'].mean()
nn_score_time = scores['score_time'].mean()
nn_accuracy = scores['test_accuracy'].mean()
nn_precision = scores['test_precision_macro'].mean()
nn_recall = scores['test_recall_macro'].mean()
nn_f1 = scores['test_f1_weighted'].mean()
nn_roc = scores['test_roc_auc'].mean()
scores
```

```python
scoring = ['accuracy', 'precision_macro', 'recall_macro' , 'f1_weighted', 'roc_auc']
scores = cross_validate(dt, x_test, y_test, scoring=scoring, cv=20)
dt_fit_time = scores['fit_time'].mean()
dt_score_time = scores['score_time'].mean()
dt_accuracy = scores['test_accuracy'].mean()
dt_precision = scores['test_precision_macro'].mean()
dt_recall = scores['test_recall_macro'].mean()
dt_f1 = scores['test_f1_weighted'].mean()
dt_roc = scores['test_roc_auc'].mean()
scores
```

```python
models_initial = pd.DataFrame({
    'Model'       : ['Gaussian NB', 'K-neighbor', 'MLP', 'decision trees'],
    'Fitting time': [gs_fit_time, kn_fit_time, nn_fit_time, dt_fit_time],
    'Scoring time': [gs_score_time, kn_score_time, nn_score_time, dt_score_time],
    'Accuracy'    : [gs_accuracy, kn_accuracy, nn_accuracy, dt_accuracy],
    'Precision'   : [gs_precision, kn_precision, nn_precision, dt_precision],
    'Recall'      : [gs_recall, kn_recall, nn_recall, dt_recall],
    'F1_score'    : [gs_f1, kn_f1, nn_f1, dt_f1],
    'AUC_ROC'     : [gs_roc, kn_roc, nn_roc, dt_roc],
    }, columns = ['Model', 'Fitting time', 'Scoring time', 'Accuracy', 'Precision', 'Recall', 'F1_score', 'AUC_ROC'])

models_initial.sort_values(by='Accuracy', ascending=False)
```

Gaussian Naive Bayes would be the best model to choose from the options tested as it gives more accurate results and better test scores than rest except MLP classifier. It takes much less time than MLP which puts it in a superior position for quicker results.
